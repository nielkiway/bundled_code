```python
import os 
from PIL import Image
import pandas as pd
import numpy as np
```

```python
data = {'ZP': [1, 2, 3, 4, 5, 6, 7, 8, 9],
        'maxSlice': [1589, 1529, 1589, 1589, 1589, 1589, 1589, 1589, 1589]}
slice_numbers = pd.DataFrame(data)
```

```python
slice_numbers.at[0,"maxSlice"]
```

```python
for zp_num in range(2,10):
    
    for slice_num in range(1,slice_numbers.at[zp_num-1, 'maxSlice']):   #range(1,1589):

        num_below = "{0:0=5d}".format(slice_num-1)
        num_current = "{0:0=5d}".format(slice_num)
        num_above = "{0:0=5d}".format(slice_num+1)

        img_path_1 = '/home/jan/Documents/Diplomarbeit/Code_zusammengefasst/QM-Meltpool-Datenaufbereitung/grayscale_area_images_all_slices/area_ZP{}_Slice'.format(zp_num) + num_below + '.png'
        img_path_2 = '/home/jan/Documents/Diplomarbeit/Code_zusammengefasst/QM-Meltpool-Datenaufbereitung/grayscale_area_images_all_slices/area_ZP{}_Slice'.format(zp_num) + num_current + '.png'
        img_path_3 = '/home/jan/Documents/Diplomarbeit/Code_zusammengefasst/QM-Meltpool-Datenaufbereitung/grayscale_area_images_all_slices/area_ZP{}_Slice'.format(zp_num) + num_above + '.png'

        img1 = Image.open(img_path_1)
        img2 = Image.open(img_path_2)
        img3 = Image.open(img_path_3)

        array1 = np.asarray(img1)
        array2 = np.asarray(img2)
        array3 = np.asarray(img3)

        new_array = np.zeros((770,775,3))
        new_array[:,:,0] = array1
        new_array[:,:,1] = array2
        new_array[:,:,2] = array3

        img = Image.fromarray(new_array.astype(np.uint8), 'RGB')
        img.save('/home/jan/Documents/Diplomarbeit/Code_zusammengefasst/QM-Meltpool-Datenaufbereitung/stacked_RGB/area_ZP{}_Slice'.format(zp_num) + num_current + '.png')
```
