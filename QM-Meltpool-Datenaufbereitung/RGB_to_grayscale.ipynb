```python
import os 
from PIL import Image
import numpy as np
```

```python
# first step: getting list with all the images in the folder 
d = '/home/jan/Documents/Diplomarbeit/Code_zusammengefasst/QM-Meltpool-Datenaufbereitung/segmented_RGB_area_sorted/standard_orig/porosity'
dest_d = '/home/jan/Documents/Diplomarbeit/Code_zusammengefasst/QM-Meltpool-Datenaufbereitung/segmented_RGB_area_sorted/standard_orig_grey/porosity'
cur_paths = []
for path in os.listdir(d):
    full_path = os.path.join(d,path)  # adding the full path as the label is stored in the path
    cur_paths.append(full_path)

# second step: transfering RGBA to grayscale + copying into new folder with grayscale pictures
for img_path in cur_paths:
    img = Image.open(img_path).convert('L')
#   #raw_array = np.asarray(img)
    name = img_path.split(os.path.sep)[-1]
    img.save(dest_d + '/' + name)
```
